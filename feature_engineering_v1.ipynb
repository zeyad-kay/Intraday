```python
import pandas as pd
```

```python
df = pd.read_csv('spy.csv')
df.rename(columns={'Unnamed: 0':'period'}, inplace = True)
df.tail()
```

```python
df.movement = df.movement.shift(-1)
df.chg = df.chg.shift(-1)
df.columns = df.columns + "_1"
df = df.rename(columns={'period_1':'period', 'movement_1': 'y'})
df
```

```python
historical = df.copy()
historical = historical.shift(1).reset_index(drop=True)
historical.columns = historical.columns + "_2"
```

```python
data = pd.concat([df, historical], axis = 1).drop(columns=['chg_1', 'period_2']).dropna().set_index('period')
```

```python
data['rvol'] = data.total_volume_1 / data.total_volume_1_2
data['rvpoc'] = data.vpoc_1 / data.vpoc_1_2
```

```python
data.to_csv('spy_train_set.csv')
```

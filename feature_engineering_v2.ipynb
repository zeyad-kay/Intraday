```python
import pandas as pd
```

```python
df = pd.read_csv('spy.csv')
df15 = pd.read_csv('15min_spy.csv',index_col=0)
df
```

```python
df15
```

```python
df.iloc[:,3:] = df.iloc[:,3:].shift(1,fill_value=-1)
df = df.rename(columns={c: c+"_prev_period" for c in df.columns if c not in ['period','movement','chg']})
df.insert(2,'movement_prev_period',df.movement.shift(1, fill_value=-1))
df.insert(4,'chg_prev_period',df.chg.shift(1))
df['movement'] = df15.movement.values
df.rename(columns={'movement':'movement_15_min'})
df
```

```python
df['rvol'] = df['total_volume_prev_period'] / df['total_volume_prev_period'].shift(1)
df['rvpoc'] = df['vpoc_prev_period'] / df['vpoc_prev_period'].shift(1)
df = df[~df['period'].str.contains('09:30:00')]
df.dropna(inplace=True)
```

```python
df = df.drop(columns=['chg']).set_index('period')
df.to_csv('spy_engineered_v2.csv')
df
```

```python
import pandas as pd
import yfinance as yf
import pandas_datareader.data as pdr
```

```python
yf.pdr_override()

data = pdr.get_data_yahoo('SPY', start='2024-03-26' ,end='2024-04-30', interval='5m')
```

```python
data
```

```python
data['shifted'] = data.Open.shift(-3)
data['movement']= (data.Open.shift(-3) > data.Open).replace({True:1, False: 0})
data.to_csv("15min_spy.csv",mode='a',header=False)
```
